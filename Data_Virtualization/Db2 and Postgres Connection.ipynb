## Adding Postgres and DB2 Connections to CP4D

### Overview and Prerequisites

We will be working with Airline data for ~12k flights. The first dataset contains flight information (FLIGHT_ID, MONTH, DAY, DAY_OF_THE_WEEK, DEPARTURE_DELAY TAXI_OUT, DISTANCE, DELAYED, YEAR). The second dataset contains information about origin and destination airports for each flight (FLIGHT_ID, ORIGIN_AIRPORT, DESTINATION_AIRPORT).  The flight information datasets have already been uploaded to the DB2 instance on the CP4D cluster and the airport information datasets have been uploaded to a Postgres database.  In this notebook, you will learn how to connect to the DB2 and Postgres databases that contain these datasets.


#### Prerequisites
1. This notebook assumes that you have a Cloud Pak for Data instance with DB2. 

### Add a Connection for Db2 on CP4D 

#### 1. Starting on the Cloud Pak for Data home page, navigate to the Sandwich icon at the top left corner, then go to 'Data' > 'Databases'. This should take you to your enabled databases.

#### 2. Click on the verticle dots next to the DB2 database and navigate to 'Details'. From this page you will need to record:
    1. Database Name: BLUDB
    2. Hostname: <hostname>
    3. Port: <port>
    4. User: <user>
    5. Password: <should be provided by database/cluster admin>
    
    Values not recorded for privacy. 
    
   
   
   
   
#### 3. Click on the Sandwich icon at the top left and naviagte 'Platform Connections'. This will take you to the `Create Connections` UI. 



#### 4. Click on 'New Connection +'. This will take you to the databases page.  Scroll down and select 'DB2'. 



#### 5. In the 'New Db2 connection' field prompts, fill in the following information: 

    1. Name: FlightInfoCP4D
    2. Database: BLUDB
    3. Hostname: collected in step 2
    4. Port: collected in step 2
    5. User: collected in step 2
    6. Password: collected in step 2
   
    5a. Make sure 'Port is SSL-enabled (optional)' is checked OFF
    
   


#### 6. Click 'Test' to test the connection and then click 'Create'. You will be able to see the connection in the 'Platform Connections' page once this step is completed successfully.



### Add a Connection for Postgres on CP4D 

#### 1. Click on the Sandwich icon at the top left and navigate to 'Platform Connections'. This will take you to the `Create Connections` UI. 



#### 2. Click on 'New Connection +'.  This will take you to the databases page.  Scroll down and select 'PostgreSQL'. 



#### 3. In the 'New PostgreSQL connection' field prompts, fill in the following information using the Postgres DB credentials that were provided: 

    1. Name: Postgres Connection
    2. Database: Get from provided DB credentials
    3. Hostname: Get from provided DB credentials
    4. Port: Get from provided DB credentials
    5. User: Get from provided DB credentials
    6. Password: Get from provided DB credentials
    7. Make sure 'Port is SSL-enabled (optional)' is checked OFF
    
#### 4. Click 'Test' to test the connection and then click 'Create'. You will have completed this step once you see the connection in the 'Platform Connections' page.
