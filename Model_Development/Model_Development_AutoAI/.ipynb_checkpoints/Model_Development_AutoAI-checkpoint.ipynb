## Developing Model using AutoAI

#### In this step, we will be using the dataset we prepared in the previous step.

#### This step can be simple or complex depending on the requirement of the usecase/model. It can involve various substeps like feature transformation from strings to numeric, creating extra mathematical features, splitting the data into training and test datasets, feature selection, iterations with multiple algorithms, and the cross validation of said iterations. 

#### In this tutorial we will show how to use AutoAI to complete all of the aforementioned substeps.
<br>
<br>

#### 1. Begin by navigating to your `Project` and selecting `Add to Project` and then `AutoAI experiment` from the menu. This will open a page where you have to provide name of the experiment. Give it a name, then click Create.


<br>
<br>
<br>
<br>

####  2. In the next screen, click `Select from project` and locate `train_flights_jan_2015.csv_shaped`. 


<br>
<br>
<br>
<br>

#### 3. We'll now choose the column we're looking to predict. In our case, we're trying to predict which flights are likely to be delayed based on certain input features. Select `Delayed`. 


<br>
<br>
<br>
<br>

####  4. Select `Experiment settings` and set a Training data split of 85% and Holdout data split of 15% by adjust the slide bar.  You can also navigate to `Prediction` in left pane and select the type of prediction (shown below). For our 

#### airline data use case, we will leave as the default `Binary classification`. Click `Save settings`.






<br>
<br>
<br>
<br>


<br>
<br>
<br>
<br>

#### 5. This will take you back to the previous screen. There click on `Run Experiment`. That will start the AutoAI experiment (as shown in the image below).


<br>
<br>
<br>
<br>

#### 6. You can click on any pipeline to get an enhanced view of the algorithm. View Evaluation metrics (i.e., ROC Curve and Confusion Matrix) and view some feature transformations dynamically created by AutoAI during the pipeline building.
#### <font color = 'red'>Note: If eight pipelines are creating problem for you while running based on cluster resource constraints, go back and change it to 4 pipelines in Experiment Settings and run it again.</font>


<br>
<br>
<br>
<br>


<br>
<br>
<br>
<br>

#### 7. After the experiments complete, save a model of your choice below to be used going forward. You can compare the performance metrics of your models created by clicking on the `Pipeline comparison` tab. Click `Save as` as shown below.


<br>
<br>
<br>
<br>

#### 8. Give it a name of your choice (or leave default) and click `Create`.

#### 9. After Model is saved go back to Asset Tab by clicking on the previous path at the top of your screen. There, you can scroll down to the `Model` section to find your saved model.

#### You also have the option to save the model as a notebook by clicking `Save as Notebook`. This can also be found on the Assets page. The notebook will showcase the code used to build the pipeline. You have the liberty to modify the code based on your preferences.

####  After changing the notebook or customizing it based on your needs/use case, you can save the model from the notebook for later deployment.
